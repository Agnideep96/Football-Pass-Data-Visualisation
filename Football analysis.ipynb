```python
#Taking the subset of the df_events for plotting purpose
df_plot_pass=df_events[['Passer','Receiver','Start_X','Start_Y','End_X','End_Y','minute']]
```

# PLOTTING THE GRAPH (Heat map)

Creating the pitch

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def create_pitch(ax):
    ax.set_facecolor("black")  # Set the field background to black

    # Pitch outline & center line
    ax.add_patch(patches.Rectangle((0, 0), 120, 80, fill=False, edgecolor="white", lw=2))
    plt.plot([60, 60], [0, 80], color="white", lw=2)

    # Penalty areas
    ax.add_patch(patches.Rectangle((0, 18), 18, 44, fill=False, edgecolor="white", lw=2))
    ax.add_patch(patches.Rectangle((102, 18), 18, 44, fill=False, edgecolor="white", lw=2))

    # 6-yard boxes
    ax.add_patch(patches.Rectangle((0, 30), 6, 20, fill=False, edgecolor="white", lw=2))
    ax.add_patch(patches.Rectangle((114, 30), 6, 20, fill=False, edgecolor="white", lw=2))

    # Circles and spots
    ax.add_patch(plt.Circle((60, 40), 10, color="white", fill=False, lw=2))  # center circle
    ax.add_patch(plt.Circle((60, 40), 0.8, color="white"))  # center spot
    ax.add_patch(plt.Circle((12, 40), 0.8, color="white"))  # left pen spot
    ax.add_patch(plt.Circle((108, 40), 0.8, color="white"))  # right pen spot

    # Arcs
    ax.add_patch(patches.Arc((12, 40), 20, 20, theta1=308, theta2=52, color="white", lw=2))
    ax.add_patch(patches.Arc((108, 40), 20, 20, theta1=128, theta2=232, color="white", lw=2))

    ax.set_xlim(0, 120)
    ax.set_ylim(0, 80)
    ax.set_aspect('equal')
    ax.axis('off')
```

The Function : plot_pass_heatmap()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects  # Needed for text outline

def plot_pass_heatmap(df, title="Pass Start Location Heatmap", team_name=""):
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.patch.set_facecolor("black")  # Set figure background to black
    create_pitch(ax)  # Draw pitch

    # Plot KDE heatmap of Start_X and Start_Y
    sns.kdeplot(
        x=df['Start_X'],
        y=df['Start_Y'],
        shade=True,
        cmap="coolwarm",
        bw_adjust=1,
        ax=ax,
        thresh=0.05,
        alpha=0.8
    )

    # Add prominent team name near right goal with outline
    text = ax.text(
        118, 40, team_name,
        fontsize=17,
        color='darkred',
        va='center',
        ha='right',
        rotation=90,
        weight='bold'
    )
    text.set_path_effects([
        path_effects.Stroke(linewidth=2, foreground='white'),
        path_effects.Normal()
    ])

    # Add Title
    plt.title(title, fontsize=14, color='white')
    plt.show()

```
